```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import dataset
dataset = pd.read_csv('Salary_Data.csv')

X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
# Split the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=1)
```

```python
# Train the simple linear regression model on the training set
from sklearn.linear_model import LinearRegression
# Build the model
regressor = LinearRegression()

# Fit and train the model
regressor.fit(X_train, y_train)
```

```python
# Predict the test results
y_pred = regressor.predict(X_test)
```

```python
# Visualize the training set results
# x axis: No. of years of experience, y: Salary
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue') # plot X_train against the prediction for the training result
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualize the test set results
# x axis: No. of years of experience, y: Salary
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue') # do not need to replace X_train with X_test as the regression line is trained on X_train
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Making a single prediction
# Predict function takes a 2D array
print(regressor.predict([[12]]))
```

```python
# Get the final linear regression equation with values of the coefficients: y = b0 + b1* x
print(regressor.coef_) # b1
print(regressor.intercept_) # b0
```
